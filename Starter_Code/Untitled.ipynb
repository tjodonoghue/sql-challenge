```python
# Function to draw ERD
def create_erd():
    fig, ax = plt.subplots(figsize=(12, 8))

    # Entity Boxes
    entities = {
        "Departments": (0.2, 0.8),
        "Dept_Emp": (0.5, 0.8),
        "Dept_Manager": (0.8, 0.8),
        "Employees": (0.2, 0.4),
        "Salaries": (0.5, 0.4),
        "Titles": (0.8, 0.4)
    }

    for entity, (x, y) in entities.items():
        ax.text(x, y, entity, ha='center', va='center', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='white'))

    # Relationships
    ax.plot([0.2, 0.2], [0.7, 0.5], color="black", marker='o')  # Departments to Dept_Emp
    ax.plot([0.5, 0.5], [0.7, 0.5], color="black", marker='o')  # Departments to Dept_Manager
    ax.plot([0.2, 0.2], [0.3, 0.5], color="black", marker='o')  # Employees to Salaries
    ax.plot([0.5, 0.2], [0.5, 0.5], color="black", marker='o')  # Employees to Dept_Emp
    ax.plot([0.8, 0.2], [0.5, 0.5], color="black", marker='o')  # Employees to Dept_Manager
    ax.plot([0.8, 0.8], [0.3, 0.5], color="black", marker='o')  # Titles to Employees

    # Remove axis
    ax.set_axis_off()

    return fig

fig = create_erd()
plt.show()
```
